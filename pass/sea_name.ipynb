```python
# 匯入工具
import requests as req
import json
from bs4 import BeautifulSoup as bs
from pprint import pprint
import pandas as pd

# 開JSON文件
with open('/Users/chianlee/Desktop/disney/sea.json', 'r') as f:
    # 讀取JSON文件資料
    res = json.load(f)

#清空/建立相關欄位list
FacilityName_list =[]
FacilityName_list.clear()

# 抓出每個元素相關值
for item in res:
    
    FacilityName_list.append({
                'FacilityName': item['FacilityName'],
                'FacilityID': item['FacilityID']
            })

# 寫入 JSON 文件
with open("/Users/chianlee/Desktop/disney/sea_namelist.json", "w") as f:
    json.dump(FacilityName_list, f, indent=4)  

# 創建 DataFrame
df = pd.DataFrame()

# 加入欄位
df["FacilityName"] = FacilityName_list
df

```
